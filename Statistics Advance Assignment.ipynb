**Q1. What is the Probability density function?**

A probability density function (PDF) is a function that describes the relative likelihood for a random variable to take on a given value. It is often used to model continuous random variables.

The PDF of a random variable X is denoted by f(x). The PDF must satisfy the following conditions:

* f(x) ≥ 0 for all x
* ∫−∞∞ f(x) dx = 1

The PDF can be used to calculate the probability that X falls within a given range of values. For example, the probability that X falls between a and b is given by:

P(a ≤ X ≤ b) = ∫ab f(x) dx

The PDF can also be used to calculate the expected value and variance of X.

The expected value of X is given by:

E(X) = ∫−∞∞ xf(x) dx

The variance of X is given by:

Var(X) = E(X²) - [E(X)]²

where E(X²) is the expected value of X².

**Q2. What are the types of Probability distribution?**


There are many different types of probability distributions, each with its own unique properties. Some of the most common types of probability distributions include:

* **Normal distribution:** The normal distribution is a bell-shaped distribution that is often used to model continuous random variables. The normal distribution is characterized by its mean and standard deviation.
* **Uniform distribution:** The uniform distribution is a distribution in which all values within a given range are equally likely. The uniform distribution is often used to model discrete random variables.
* **Binomial distribution:** The binomial distribution is a distribution that describes the number of successes in a sequence of independent experiments, each of which has a constant probability of success. The binomial distribution is often used to model the number of successes in a sample of a given size.
* **Poisson distribution:** The Poisson distribution is a distribution that describes the number of events that occur in a fixed interval of time or space. The Poisson distribution is often used to model the number of phone calls received by a call center in a given hour.
* **Exponential distribution:** The exponential distribution is a distribution that describes the time until the next event occurs. The exponential distribution is often used to model the time until the next failure of a machine.

These are just a few of the many different types of probability distributions that are available. The choice of which probability distribution to use will depend on the specific problem being modeled.


**Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.**

```python

import math

def normal_pdf(mean, std_dev, x):
  """
  This function calculates the probability density function of a normal distribution with given mean and standard deviation at a given point.

  Args:
    mean: The mean of the normal distribution.
    std_dev: The standard deviation of the normal distribution.
    x: The point at which to evaluate the probability density function.

  Returns:
    The probability density function of the normal distribution at the given point.
  """

  exponent = -((x - mean)**2) / (2 * std_dev**2)
  coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))

  return coefficient * math.exp(exponent)
```

**Q7. How Binomial distribution different from Poisson distribution?**


**Binomial distribution:**

* Models the number of successes in a sequence of independent experiments, each with a constant probability of success.
* Discrete distribution.
* Parameters: number of trials (n) and probability of success (p).
* Mean: n * p
* Variance: n * p * (1 - p)

**Poisson distribution:**

* Models the number of events that occur in a fixed interval of time or space.
* discrete distribution.
* Parameter: rate parameter (λ).
* Mean: λ
* Variance: λ

**Key differences:**

* **Number of trials:** Binomial has a fixed number of trials, Poisson has an unbounded number of trials.
* **Parameterization:** Binomial has two parameters (n and p), Poisson has one parameter (λ).
* **Mean-variance relationship:** For the binomial distribution, the variance is equal to the product of the mean and the probability of failure. For the Poisson distribution, the mean and variance are equal.

**Examples:**

* Flipping a coin 10 times and counting the number of heads (binomial).
* Counting the number of phone calls received by a call center in a given hour (Poisson).

**Q10. In normal distribution with respect to mean position, where does the least frequent data appear?**

The least frequent data in a normal distribution appears in the tails of the distribution, which are located furthest away from the mean. This is because the normal distribution is bell-shaped, with the most frequent data appearing near the center of the distribution. As you move away from the center, the data becomes less frequent.
